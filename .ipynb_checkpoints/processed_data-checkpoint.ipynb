```python
def split_dataset1_train_test(name):
    dset1_ts_close = []
    dset1_tr_close = []
    dset1_ts_open = []
    dset1_tr_open = []
    if name == None:
        return None
    elif name == 'close':
        for i, d in enumerate(dset1_close): 
            if i%4 == 0:
                dset_1_ts_close.append(d)
            else:
                dset_1_tr_close.append(d)
        return dset1_ts_close, dset_1_tr_close
    
    elif name == 'open':
        for i, d in enumerate(dset1_open): 
            if i%4 == 0:
                dset_1_ts_open.append(d)
            else:
                dset_1_tr_open.append(d)
        return dset1_ts_open, dset1_tr_open 
```

```python
split_dataset1_train_test('close')
```
